```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("airbnb.csv")
df.head()
```

```python
df.info()
```

# Checking for missing data

```python
df.isnull().sum()
```

We can remove those columns which are not required. 



```python
df.drop(columns = ["license", "house_rules"], errors = "ignore", inplace=True)
df.info()
```

```python
df["last review"] = pd.to_datetime(df["last review"], errors = "coerce")
#error = coerce helps to prevent the errors. We can see that there are some missing values in last review column. Now when the function 
#tries to convert that empty data into date and time it results error. So to tackle it, coerce will convert it into NaN. 
```

```python
df.info()
```

```python
df.head(2)
```

```python
# We will fill the missing values of review per month with 0 because there might not be people to give review. 
df.fillna({"reviews per month" : 0, 
            "last review": df["last review"].min()
          }, inplace = True)
```

```python
df.dropna(subset=["NAME", "host name"], inplace = True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#removing dollar sign and convert to ffloat because that dollar sign may create problem for us.
df["price"] = df["price"].replace('[\$,]', "", regex = True).astype(float) 
df["service fee"] = df["service fee"].replace('[\$,]', "", regex = True).astype(float)
```

```python
df.head()
```

# Removing Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

# Descriptive Statistics

```python
df.describe()
```

# Visualization

What is the distribution of listing prices?

```python
plt.figure(figsize=(10,6))
sns.histplot(df["price"], bins = 50, kde = True, color = "red")
plt.title("Distribution of Listing Price") 
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

The above histogram provides that our data is evenly distributed across different prices.

The KDE line helps us to visualize the trend.

# How are room types distributed?

```python
plt.figure(figsize = (8,5))
sns.countplot(x = "room type", data = df, color = "green")
plt.title("Room Distribution")
plt.xlabel("Room Type")
plt.show()
```

From this information we can conclude that people mainly book entire home or apartment and private room from airbnb.

# How are listing distributed across neighbourhoods?

```python
df.info()
```

For this question fifth and sixth columns are necessary.

```python
plt.figure(figsize=(10,6))
sns.countplot(y = "neighbourhood group", data = df, color = "pink", order = df["neighbourhood group"].value_counts().index)
plt.title("Number of Listings by Neighbourhood Group")
plt.xlabel("Count")
plt.ylabel("Neighbourhood Group") 
plt.show()
```

From this info, we can say that company should focus on Manhattan because there are more listings from this place.

# What is the relation between price and room type?

```python
plt.figure(figsize = (10,6))
sns.boxplot(x = "room type", y = "price", hue= "room type", data = df, palette="Set1")
plt.title("Price VS Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price $")
plt.legend(title = "Room Type")
plt.show()
```

This plot says that the booking of private room ranges from 350 to 900 dollars whereas for hotel room it ranges from 500 to 850 dollars.

# How has the number of reviews changed over with time?

```python
df["last review"] = pd.to_datetime(df["last review"])
reviews_over_time = df.groupby(df["last review"].dt.to_period("M")).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind = "line", color = "red")
plt.title("Number of Reviews Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Reviews")
plt.show()
```
