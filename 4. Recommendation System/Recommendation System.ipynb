The movie in first row has these tangled data about the genres. so now we are going to extract the things that will be useful like: Action, Adventure. Also note that this is not a list nor a dict but a string.
For that we will import ast which has a function called literal_eval() which will convert a string into the list.

```python
import ast
def conversion(obj): 
    list1 =[]
    for i in ast.literal_eval(obj): 
        list1.append(i["name"]) 
    return list1 
```

```python
movies["genres"] = movies["genres"].apply(conversion) 
# this one goes through the every element of genres column and applies the conversion function and stores thoes values in the same column.
```

```python
movies.head()
# we can see that we have sorted our genres column. Now it's the turn for keywords column.
```

```python
movies.keywords = movies.keywords.apply(conversion)
```

```python
movies.head()
```

```python
# for cast column, there are numerous characters and the real name of the character played by. We only need the top three characters' real name
# of the movie. So, we are going to create a new function called cast_pickup which will pick the three main characters of the movie.

def cast_pickup(obj):
    list2 = []
    count = 0
    for i in ast.literal_eval(obj):
        if count == 3: 
            break
        else: 
            list2.append(i["name"])
            count += 1
    return list2
movies.cast = movies.cast.apply(cast_pickup)
```

```python
movies.head()
```

```python
# we can see that we have got the names of cast in a list. Now we go for crew. In crew, we will focus on to bring 
# director because people love to see the movies of a particular director.
```

```python
movies.crew.iloc[0]
```

```python
def get_director(obj): 
    list3 = [] 
    for i in ast.literal_eval(obj): 
        if i["job"] == "Director": 
            list3.append(i["name"])
            break 
    return list3 
    
```

```python
movies.crew.apply(get_director)
```

```python
movies.crew = movies.crew.apply(get_director)
```

VECTORIZATION


Now we going to do text vectorization of tags column. There are numerous techniques of coverting a text into vectors; bag of words, tf idf, wordsvec and so on. We will be using bag of words because it's basic. 

For now we will combine the tags of all the movies which will give us a very enormous string of texts. Then we will feed this very large string into a library called scikitlearn and extract the words having highest frequency. I am extracting 5000 words, you can do more than 5000 also. 
Let's suppose we have the first word called action which has the highest frequency. Now we will again go into the tags of the first movie and calculate the frequency of the word action in the tags. If there are two then we will assign 2. Similarly we will move to the next word let's say water is that word. Again we will go through the tags of the first movie and calculate the frequency of the word. We will do it for all the 5000 words and take out the data which is called a vector. And it looks like this: 
[2,0,9,0...] => first movie
[0,0,1,2,..] => second movie

But before that let me explain something. 

Why are we doing text vectorization?
We are building a movie recommendation system which will basically recommend the movies to the user based on similarity. Our model doesn't understands texts. It only understands numbers. We are going to convert the text into numbers using CountVectorizer class in scikitlearn library, based on the occurrence of the words in the movie tags. I hope you are getting my point.


```python
from sklearn.feature_extraction.text import CountVectorizer
```

Stop Words: Those words which don't have meaning in a sentence but are used in the formation of a sentence. Like: are, is , the, so on.


```python
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
# We have to explicitly convert this into array because it return us an object.
cv.fit_transform(freshdf["tags"]).toarray()
```

```python
vectors = cv.fit_transform(freshdf["tags"]).toarray()
```

```python
# let's see the vector of the first movie which is avatar.
vectors[0]
```

```python
# let's see those 5000 most frequent words. 
# feature_names = cv.get_feature_names_out()
# for word in feature_names:
#    print(word)
```

```python
#Now we have a problem. We could see three words naming: brings bringing and bring which ultimately means bring.
# Due to this, we have so many words which basically carry the almost same meaning. 
# So, now we are going to stem it. 
# What is stemming? 
# Let me give you an example: It turns live lives living into live (liv). It helps to reduce our tags. 
# We will need a library named nltk. 
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text): 
    list4 = [] 
    for i in text.split():
        list4.append(ps.stem(i))
    return " ".join(list4)
```

```python
freshdf.tags = freshdf.tags.apply(stem)
```

```python
freshdf.head()
#vectors uncomment it
```

Finally we have converted our movies tags into vectors. Now we will calculate the distance between the vectors. There are two ways: euclidean distance(x2 - x1 type) and cosine distance(angle). In higher dimension vector, cosine distance are more effective.
For this we have a function in scikitlearn named cosine_similarity.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

What we have done is that we calculated the cosine similarity among the movies and stored it in similarity which is a array of lists. The function bascially takes the first movie and calculated the cosine similarity with all the other remaining 4085 movies and stores it in a list. Then it takes second movie and does the same thing. 
If we look at the first list which is [1.        , 0.08346223, 0.0860309 , ..., 0.04499213, 0.        ,
        0.        ]. We have 1 at first index because obviously first movie has 100% similarity with 1st movie. Then we have 0.08346, which is the comparison between first and second movie. 

If we sort the similarities value then we will lose the movie index. So for that we will use enumerate. Then we will sort on the basis of the value.

```python
enumerate(similarity[0])
```

```python
#list(enumerate(similarity[0]))
```

This is a list of tuples with movie index and similarity value.

```python
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x: x[1] ) [1:6]
```

Now we will embbed this code into the function so that it works for every movie.

```python
def recommend(movie): 
    movie_index = freshdf[freshdf["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_5list = sorted(list(enumerate(distances)), reverse=True, key = lambda x: x[1] ) [1:6]

    for i in movies_5list: 
        print(freshdf.iloc[i[0]].title)  
```

```python
recommend("Avatar")
```

```python
recommend("Batman Begins")
```
